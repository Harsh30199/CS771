```python
import numpy as np
import time
from IPython.display import display, Latex ##Used to print latex formatted text

import matplotlib.pyplot as plt

#Sets the plots to be displayed and saved in notebook
%matplotlib inline 
plt.rcParams["figure.figsize"] = (20,10)
```

# Q1(a)

```python
def gradient_descent(gradient,init_,learn_rate,n_iter=50, tol=1e-06):
    x = init_  ##Starting point for gradient descent algorithm
    for i in range(n_iter): 
        delta = -learn_rate * gradient(x) ##Change in x directon is calculated
        if np.all(np.abs(delta)<=tol): ##if change is less than thresh break out of loop
            break
        x += delta ##New point x calculated
    return round(x*1000)/1000 ##Return final x value calculated
```

### Gradient function is first differential of function to optimise
### init_ is chosen by trial and error

### Minima for $x^2 + 3x + 4$

```python
min_x = gradient_descent(lambda x: 2*x + 3 , init_ = 0.0 , learn_rate = 0.1)
min_value = (pow(min_x,2))+(3*min_x)+4

display(Latex('Minima for ' + '$ x^2 + 3x + 4$'+ f' is at x = {min_x} with value {min_value}'))

```

### Minima for $x^4 - 3x^2 + 2x$

```python
min_x = gradient_descent(lambda x: 4*pow(x,3)-(6*x)+2 , init_ = 0.0 , learn_rate = 0.01,n_iter =100)
min_value = (pow(min_x,4))-(3*pow(min_x,2))+(2*min_x)

display(Latex('Minima for ' + '$ x^4 - 3x^2 + 2x$'+ f' is at x = {min_x} with value {min_value}'))
```

# Q1(b)

```python
def gradient(X,Y,a,b):
    pred_y = a*X + b ##Creates a array of predicted y value
    n = len(X) ##No of observation
    grad_a = (-2/n) * sum(X * ( Y - pred_y )) ##Calculates gradient of a
    grad_b = (-2/n) * sum(Y - pred_y)  ##Calculates gradient of b
    
    return (grad_a,grad_b)
```

```python
def gradient_descent2(X,Y,init_a,init_b,learn_rate,n_iter=50, tol=1e-05):
    a = init_a ##Starting point of a for gradient descent algorithm
    b = init_b ##Starting point of a for gradient descent algorithm
    for i in range(n_iter): 
        delta = -learn_rate * np.array(gradient(X,Y,a,b)) ##Calculates values by which a and b have to be updated 
        if np.all(np.abs(delta)<=tol): ##if change is less than thresh break out of loop
            break
        a += delta[0] ##Updates a in opposite direction of gradient
        b += delta[1] ##Updates b in opposite direction of gradient
    a = round(a*1000)/1000
    b = round(b*1000)/1000
   
    return (a,b) ##Return final a,b value calculated

```

# Q1(c)

```python
np.random.seed(0) ##Initial point for random number generation

X = 2.5 * np.random.randn(10000) + 1.5 ##Generates 10000 values with mean = 1.5 and standard deviation = 2.5
res = 1.5 * np.random.randn(10000) ##Generates 10000 residual terms 
Y = 2 + 0.3 * X + res ##Generates actual value of Y 

start_time = time.time()##Time just before execution of following statement
a,b = gradient_descent2(X,Y,0,0,0.001,10000) ##Calculates optimal parameters (a,b) relating X with Y
print("Execution Time: %s seconds " % (time.time() - start_time))##Calculates time required for execution

print('a =',a,', b =',b)
```

### Value for a = 0.297 and b = 2.016 is very near to expected values of a = 0.3 and b = 2.0

# Q1(d)

```python
def gradient_descent3(X,Y,init_a,init_b,learn_rate,n_iter=50,batch_size=100, tol=1e-05):
    a = init_a ##Starting point of a for gradient descent algorithm
    b = init_b ##Starting point of a for gradient descent algorithm
    

    for i in range(n_iter):
        
        index = np.arange(0,len(X)) ##Generates index values from 0 to len(X) - 1 
        np.random.shuffle(index) 
        m = [X[index[0:batch_size]],Y[index[0:batch_size]]]
        delta = -learn_rate * np.array(gradient(m[0],m[1],a,b)) ##Calculates values by which a and b have to be updated 
        if np.all(np.abs(delta)<=tol): ##if change is less than thresh break out of loop
            break
        a += delta[0] ##Updates a in opposite direction of gradient
        b += delta[1] ##Updates b in opposite direction of gradient
    a = round(a*1000)/1000
    b = round(b*1000)/1000
    
    
    return (a,b) ##Return final a,b value calculated
```

```python

start_time = time.time()##Time just before execution of following statement
a,b = gradient_descent3(X,Y,0,0,0.001,10000,500) ##Calculates optimal parameters (a,b) relating X with Y
print("Execution Time: %s seconds " % (time.time() - start_time))##Calculates time required for execution

print('a =',a,', b =',b)
```

# Q1(e)

### To check if MiniBatch SGD and SGD performs better or worse in respect to time required for execution, we will run MiniBatch SGD and SGD 10 times and will take average of execution time for each

```python
total_time =[]
for i in range(10):
    start_time = time.time()##Time just before execution of following statement
    a,b = gradient_descent3(X,Y,0,0,0.001,10000,500) ##Calculates optimal parameters (a,b) relating X with Y
    t = time.time() - start_time ##Calculates time required for execution
    total_time.append(t)
    print('Time taken in Iteration ',i,' :',t)
    print('a =',a,', b =',b)

print("Average Time taken by MiniBatch SGD :" , np.sum(total_time)/10)
```

### MiniBatch SGD does better as it is observed that average execution time of MiniBatch Stochastic Gradient Descent is always very less compared to that of Gradient Descent for same parameter initialisation and learning rate

```python
total_time =[]
for i in range(10):
    start_time = time.time()##Time just before execution of following statement
    a,b = gradient_descent3(X,Y,0,0,0.001,10000,1,1e-10) ##Calculates optimal parameters (a,b) relating X with Y
    t = time.time() - start_time ##Calculates time required for execution
    total_time.append(t)
    print('Time taken in Iteration ',i,' :',t)
    print('a =',a,', b =',b)
    

print("Average Time taken by SGD :" , np.sum(total_time)/10)
```

### SGD does better as it is observed that average execution time of  Stochastic Gradient Descent is always  less than to that of Gradient Descent for same parameter initialisation and learning rate but different gradient threshold, thresh for SGD is $10^{-10}$  while of Gradient Descent thresh is $10^{-5}$

### To check if Optimal Batch Size exists we will test Batch Size from 100 to 1000 in increements of 50, 10 times over dataset, storing parameter (a,b) values predicted , Root Mean Square Testing Error  and ExecutionTime

```python
##Function to calculate Root Mean Square Error 
def rmse(X,Y,a,b):
    predy = a*X + b
    error = (((sum((predy-Y)**2))/len(X))**0.5)
    return error
```

```python
error={}##Stores average RMSE for each Batch Size
timeexec={}##Average Time required for execution Batch Size wise

##Split Dataset into training and test set in ratio 9:1
index = np.arange(0,10000) ##Generates index values from 0 to 9999 
np.random.shuffle(index)   ##Shuffles the index list
X_test = X[index[0:1000]]  ##First 1000 values become test set
Y_test = Y[index[0:1000]]  ##First 1000 values become test set
X_train = X[index[1000:]]  ##Rest values become train set
Y_train = Y[index[1000:]]  ##Rest values become train set

for bsize in range(100,1001,50):
    e=[]
    list_t=[]
    for i in range(10):
        start_time = time.time()##Time just before execution of following statement
        a,b = gradient_descent3(X_train,Y_train,0,0,0.001,10000,bsize) ##Calculates optimal parameters (a,b) relating X with Y
        t = time.time() - start_time ##Calculates time required for execution
        e.append(rmse(X_test,Y_test,a,b))##Calculates RMSE
        list_t.append(t)
        
    error[bsize] = sum(e)/10 ##Average RMSE
    timeexec[bsize]=sum(list_t)/10 ##Average execution time
```

```python
fig , ax = plt.subplots(1)
ax.set_xlabel('Batch Size')
ax.set_ylabel('Time')
ax.set_title('Average Execution Time V/S Batch Size')
keys = list(timeexec.keys())
values = list(timeexec.values())
plt.locator_params(axis='x', nbins=20) 
ax.plot(keys, values)
print('Minimum Execution time of',min(values),'seconds for Batch Size of',keys[values.index(min(values))])
print('Differnce in Max Execution time and Min Execution time is',max(values)-min(values),'seconds')
```

```python
fig , ax = plt.subplots(1)
ax.set_xlabel('Batch Size')
ax.set_ylabel('RMSE')

ax.set_title('RMSE V/S Batch Size')
keys = list(error.keys())
values = list(error.values())
plt.locator_params(axis='x', nbins=20) 
ax.plot(keys, values,color='red')

print('Difference in Max RMSE error and Min RMSE error is',max(values)-min(values))
```

## INFERENCES

### If we choose Optimal Batch Size based on time required to predict parameter values then in general Smaller Batch Size gives better performance based on time than larger Batch Size as can be seen in above Average Execution Time V/S Batch Size Graph

### If we choose Optimal Batch Size based on Average Root Mean Square Error (RMSE) then there is no definite Optimal Batch Size as can be seen in the above RMSE V/S Batch Size Graph as  difference between Max RMSE and Min RMSE is of order $ 10 ^ {-3}$ which is not very significant in comparison to RMSE  which is in range 1.4 to 1.5
