### Derivation of MLE for a Normal Distribution <br><br>

$\large{\text{ Probability Density Function (PDF) of Normal Distribution :  }}$ <br><br>

$$\large{f(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}}$$ <br><br>

$\large{\text{ Suppose we have } n \text{ independently and identically distributed observations as } x_{1},x_{2},\dots,x_{n} }$ <br><br>

$\large{\text{ As they are independent, Likelihood  given } \mu \text{ and } \sigma \text{ is }}$ <br><br>


$$\large{\begin{align}
L(x_{1},x_{2},\dots,x_{n}|\sigma,\mu)&=\prod_{i=1}^{n}\frac{1}{\sigma\sqrt{2\pi}} \quad \mathbf{exp \Bigg\{} -\frac{1}{2\sigma^{2}} \bigl({x_{i} - \mu } \bigr)^{2} \mathbf{\Bigg\}}\\ \\
&=\Biggl(\frac{1}{\sigma\sqrt{2\pi}}\Biggr)^{n} \quad \mathbf{exp \Bigg\{} -\frac{1}{2\sigma^{2}} \sum_{i=1}^{n}\bigl({x_i - \mu} \bigr)^2 \mathbf{\Bigg\}} \\\\
&\propto \mathbf{exp \Bigg\{} -\frac{1}{2\sigma^{2}} \sum_{i=1}^{n}\bigl({x_i ^ 2  + \mu ^ 2 - 2 x_i \mu } \bigr)\mathbf{\Bigg\}} \\\\
&= \mathbf{exp \Bigg\{} -\frac{1}{2\sigma^{2}} \left(\sum_{i=1}^{n}x_i ^ 2  + n \mu ^ 2 - 2 n \bar{x} \mu  \right)\mathbf{\Bigg\}} \\\\
&= \mathbf{exp \Bigg\{} -\frac{n}{2\sigma^{2}} \left(\mu ^ 2 - 2 \bar{x} \mu + \frac{1}{n} \sum_{i=1}^{n}x_i ^ 2 \right)\mathbf{\Bigg\}} \ldots\ldots (1) \\\\
\end{align}}$$
<br><br>

$\large {\text{ As we know Likelihood is also Normal, equation in form of } \mu_{MLE} \text{ and } \sigma_{MLE} \text{ is }}$
<br><br>

$$\large{\begin{align}
&=\Biggl(\frac{1}{\sigma_{MLE}\sqrt{2\pi}}\Biggr) \quad \mathbf{exp \Bigg\{} -\frac{1}{2\sigma_{MLE}^{2}} \bigl({\mu - \mu_{MLE}} \bigr)^2 \mathbf{\Bigg\}} \\\\
&\propto \mathbf{exp \Bigg\{} -\frac{1}{2\sigma_{MLE}^{2}} \bigl({\mu - \mu_{MLE}} \bigr)^2 \mathbf{\Bigg\}} \\\\
&= \mathbf{exp \Bigg\{} -\frac{1}{2\sigma_{MLE}^{2}} \bigl({\mu ^ 2  + \mu_{MLE} ^ 2 - 2 \mu \mu_{MLE} } \bigr)\mathbf{\Bigg\}}\ldots\ldots (2) \\\\
\end{align}}$$
<br><br>

$\large{\text{Comparing Coefficents of } \mu_{MLE} \text{ and } \sigma_{MLE} \text{ in (1) and (2) , we get } }$
<br><br>

$$\large{\mu_{MLE} = \bar{x}}$$<br><br>
$$\large{\sigma_{MLE} = \frac{\sigma}{\sqrt{n}}}$$ <br><br>

$\large{\text{Solving above two equations}}$
<br><br>

$$\large{\frac{1}{2\sigma^{2}}\sum_{i=1}^{n}(2\hat{\mu}-2x_{i})=0}$$ <br><br>

$$\large{\mu_{MLE}=\frac{\sum_{i=1}^{n}x_{i}}{n}}$$ <br><br>

$\large{\text{Differentitating } L \text{ with respect to } \sigma}$
<br><br>

$$\large{\frac{d\mathcal{L}}{d\sigma}=-\frac{n}{\sigma}+\sum_{i=1}^{n}(x_{i}-\mu)^{2}\sigma^{-3}}$$
<br><br>

$$\large{\frac{d\mathcal{L}}{d\sigma}= 0 }$$ <br><br>

$\large{\text{Solving above two equations}}$
<br><br>

$$\large{\sigma_{MLE}^{2}=\frac{\sum_{i=1}^{n}(x_{i}-\hat{\mu})^{2}}{n}}$$<br><br>

$\large{\text{Standard Deviation }(\sigma_{MLE}) \text{ is } }$
<br><br>

$$\large{\sigma_{MLE}=\sqrt{\frac{\sum_{i=1}^{n}(x_{i}-\hat{\mu})^{2}}{n}}}$$<br><br>

### As we know that Likelihood follows Normal Distribution, then knowing $ \mu_{MLE}$ and $\sigma_{MLE} $ we can easily estimate Probability Distribution Function of Likelihood to be 

### $$L \sim \mathcal{N}(\mu_{MLE},\,\sigma_{MLE}^{2}) $$

### Where $\mu_{MLE}$ is mean of observed data and $\sigma_{MLE}$ is standard deviation of observed data 

```python
def normcalc(x,mu,std):
    
    return(1/(std * np.sqrt(2 * np.pi)) *np.exp( - (x - mu)**2 / (2 * std**2)))
```

```python
mu_likelihood = np.mean(data)
std_likelihood = np.sqrt(25/50) ##np.std(data) 

mu_space1 = np.linspace(mu_likelihood - 4* std_likelihood,mu_likelihood + 4* std_likelihood,1000)
likelihood=[]
for i in mu_space1:
    likelihood.append(normcalc(i,mu_likelihood,std_likelihood))

mu_space2 = np.linspace(25-4*5,25+4*5,1000)
prior =[]
for i in mu_space2:
    prior.append(normcalc(i,25,5))
 
```

$$\large{\sigma_{MAP} = \sqrt{\biggl(\frac{1}{\sigma_{prior}^2} + \frac{n}{\sigma_{data}^2}\biggr)^{-1}}} $$<br><br>


$$\large{\mu_{MAP} = \sigma_{MAP}^2 \biggl(\frac{\mu_{prior}}{\sigma_{prior}^2} +  \frac{n\bar{x}}{\sigma_{data}^2}\biggr)}$$ <br><br>

$$\large{\sigma_{prior}\text{ is standard deviation of prior}}$$<br>
$$\large{\mu_{prior}\text{ is mean of prior}}$$<br>
$$\large{\bar{x}\text{ is mean of  observed data}}$$<br>
$$\large{\sigma_{data}\text{ is standard deviation of data}}$$<br>
$$\large{n\text{ is No of Observations}}$$<br><br>

### As we know that Posterior follows Normal Distribution, then knowing $ \mu_{MAP}$ and $\sigma_{MAP} $ we can easily estimate Probability Distribution Function of Posterior to be 

### $$P \sim \mathcal{N}(\mu_{MAP},\,\sigma_{MAP}^{2}) $$


```python
std_posterior = np.sqrt(1 / ((1/25) + (1/(25/50))))
mu_posterior = (std_posterior**2)*((25/25) + (mu_likelihood*50/25))

mu_space3 = np.linspace(mu_posterior - 4* std_posterior,mu_posterior + 4* std_posterior,1000)
posterior = []
for i in mu_space3:
    posterior.append(normcalc(i,mu_posterior,std_posterior))
    
plt.plot(mu_space1,likelihood, color = 'b',label= 'likelihood')
plt.plot(mu_space2,prior , color = 'r', label = 'prior')
plt.plot(mu_space3,posterior, color = 'g' , label = 'posterior')
plt.legend(loc = 'upper right')
```

```python

def mcmc(hops,mu,sigma):
    proposed_posterior = []
    current_mu = mu
    for i in range(hops):
        proposed_posterior.append(current_mu)
        proposed_mu = np.random.normal(current_mu,sigma)
        
        curr_prob = normcalc(current_mu,mu_likelihood,std_likelihood) * normcalc(current_mu,25,5)
        move_prob = normcalc(proposed_mu,mu_likelihood,std_likelihood) * normcalc(proposed_mu,25,5)
        
        r = move_prob/curr_prob
        
        if r > 1 or (r <= 1 and r > np.random.uniform(0,1)):
            current_mu = proposed_mu
    return proposed_posterior
```

```python
proposal_posterior = mcmc(100000,mu=0,sigma=1)
plt.hist(proposal_posterior,bins=100,density = True) 
plt.plot(mu_space3,posterior, color = 'r' , label = 'Analytical Posterior',linewidth = 3)
plt.legend(loc = 'upper right')
```
